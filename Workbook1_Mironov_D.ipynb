**Пример 4.2.2**

Для функции sin(𝑥𝑒cos(𝑥)) на интервале [0;10] с шагом 0,05 построить
график функции, закрасить площадь под ним и вычислить значение
этой площади.
Для вычисления площади воспользуйте функции trapz(), вычисляющей
интеграл по правилу трапеции. Для ее корректной работы необходимо
подключить следующие библиотеки:
from scipy.integrate import simps
from numpy import trapz


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simps
from numpy import trapz

x = np.arange(0.0, 10, 0.1)
y = np.abs(np.sin(x * np.exp(np.cos(x))))
plt.grid()
plt.plot(x, y, c='r')
plt.fill_between(x, y)

area = trapz(y)
print(area)
```

**Задание 4.3.3**

Дана функция |cos(𝑥𝑒(cos(𝑥)+ln(𝑥+1)))| Построить график на
интервале (0,10) с шагом 1 с заливкой площади и найти эту площадь
под ним. Для вычисления площади воспользуйте функции trapz(),
вычисляющей интеграл по правилу трапеции.
Для ее корректной работы необходимо подключить следующие
библиотеки:
from scipy.integrate import simps
from numpy import trapz


```python
import numpy as np
from matplotlib import pyplot as plt
from numpy import exp, cos, log, abs, trapz
def f(x):
    return abs(cos(x * exp(cos(x) + log(x + 1))))
x = np.arange(0, 10, 1)
y = f(x)
plt.grid()
plt.plot(x, y)
plt.fill_between(x, y)
plt.show()
area = trapz(y)
print('area:', area)
```

**Задание 4.3.5**

Создайте простейший калькулятор, включающий основные действия
для двух переменных ′ + ′, ′ − ′, ′ ∙ ′, ′ ÷ ′, а также вычисление
следующих функций: 𝑒𝑥+𝑦, sin(𝑥 + 𝑦) , cos(𝑥 + 𝑦) , 𝑥𝑦.

```python
from math import exp, sin, cos
def add(a, b):
    return a + b
def subtract(a, b):
    return a - b
def divide(a, b):
    if b == 0:
        raise ZeroDivisionError("На ноль делить нельзя")
    return a / b
def multiply(a, b):
    return a * b
def expf(a, b):
    return exp(a + b)
def sinf(a, b):
    return sin(a + b)
def cosf(a, b):
    return cos(a + b)
def powf(a, b):
    return a ** b
def helpf():
    return 'Доступные операции:\n' \
           'Базовые: +, -, /, *\n' \
           'Дополнительные: sin - sin(a+b), cos - cos(a+b), exp - e^(a+b)\n'\
           'Выход: exit'
def calc(a, b, op):
    pattern = "{} {} {} = {}"
    if op == '+':
        return pattern.format(a, op, b, add(a, b))
    elif op == '-':
        return pattern.format(a, op, b, subtract(a, b))
    elif op == '*':
        return pattern.format(a, op, b, multiply(a, b))
    elif op == '/':
        return pattern.format(a, op, b, divide(a, b))
    elif op == 'exp':
        return "e^({} + {}) = {}".format(a, b, expf(a, b))
    elif op == 'sin':
        return "sin({} + {}) = {}".format(a, b, sinf(a, b))
    elif op == 'cos':
        return 'cos({} + {}) = {}'.format(a, b, cosf(a, b))
    elif op == 'help':
        return helpf()
    else:
        return None
def main():
    print(helpf())
    while True:
        operation = str(input("Введите операцию:"))
        if operation == 'exit':
            break
        a = float(input("Введите первое число:"))
        b = float(input("Введите второе число:"))
        print("Результат:", calc(a, b, operation))
if __name__ == "__main__":
    main()
```
